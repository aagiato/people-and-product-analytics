# HR EDA (IBM Attrition)
Loads data and computes overall attrition %, then by Department/JobRole.

```python
import pandas as pd, pathlib
p = pathlib.Path('people_analytics_hr/data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
if not p.exists():
    print('Download the dataset to', p)
else:
    df = pd.read_csv(p)
    df['IsAttrited'] = (df['Attrition']=='Yes').astype(int)
    overall = df['IsAttrited'].mean()
    print('Overall attrition %:', round(100*overall,2))
    by_dept = df.groupby('Department')['IsAttrited'].mean().sort_values(ascending=False)
    display(by_dept)
```
